```python
#We gaan nu de t testen weer uitvoeren. De t scores zijn gelijk (gelukkig maar :-))
#We krijgen echter veel meer informatie. Cohen d, Hedge g,Glass delta en Pearson r zijn
#scores van de effect size. We kiezen voor Pearson's r.
rp.ttest(group1= random_subset['Flight_Distance'][random_subset['satisfaction'] == 1], group1_name= "tevreden",
         group2= random_subset['Flight_Distance'][random_subset['satisfaction'] == 0], group2_name= "ontevreden")
```

```python
rp.ttest(group1= random_subset['Departure_Delay_in_Minutes'][random_subset['satisfaction'] == 1], group1_name= "tevreden",
         group2= random_subset['Departure_Delay_in_Minutes'][random_subset['satisfaction'] == 0], group2_name= "ontevreden")
```

```python
#Dat is mooie output; hoe hoger r, hoe groter de effect size. Flight distance heeft het grootste effect;
#departure delay speelt nauwelijks een rol van betekenis

#Eigenlijk moeten we telkens eerst de assumpties van een t test
#controleren. Gegeven de steekproefomvang maken we ons geen zorgen over
#de (normaal) verdeling. Belangrijk is dat er sprake is van vergelijkbare variantie in beide groepen
#Dat wordt gecheckt met Levene's test. 
stats.levene(random_subset['Flight_Distance'][random_subset['satisfaction'] == 1],
             random_subset['Flight_Distance'][random_subset['satisfaction'] == 0],
             center= 'mean')

```

```python
#Dat betekent uitwijken naar Welch's t test of een non parametrische test doen
#Sommige wetenschappers pleiten ervoor om direct te kiezen voor Welch's t test
#Als er vergelijkbare variantie is, levert deze hetzelfde resultaat op als de t test
#Als er geen vergelijkbare variantie is, corrigeert deze daarvoor
stats.ttest_ind(random_subset['Flight_Distance'][random_subset['satisfaction'] == 1],
                random_subset['Flight_Distance'][random_subset['satisfaction'] == 0], equal_var=False)
```

```python
#Het alternatief: de non parametrische toets, de Mann Whitney
stats.mannwhitneyu(random_subset['Flight_Distance'][random_subset['satisfaction'] == 1],
                random_subset['Flight_Distance'][random_subset['satisfaction'] == 0])
```

```python
#Dan is het nu tijd voor de ANOVA: we gaan drie groepen qua leeftijd met elkaar vergelijken,
#namelijk de klasse die men vliegt. Is er een verschil in gemiddelde leeftijd per klasse; de klassen
#zijn gehercodeerd (business = 1, eco plus = 2 en eco = 3)
random_subset['Class'].replace({'Business': 1, 'Eco Plus': 2, 'Eco': 3},inplace = True)

```

```python
#Eigenlijk moeten we natuurlijk eerst eens zelf de data vergelijken
Age_business_class = random_subset[random_subset['Class'] == 1]['Age']
Age_eco_plus = random_subset [random_subset ['Class'] == 2]['Age']
Age_eco = random_subset [random_subset ['Class'] == 3]['Age']

```

```python
#Gemiddelde van de drie groepen uitrekenen
print(np.mean(Age_business_class))
print(np.mean(Age_eco_plus))
print(np.mean(Age_eco))
```

```python
#Standaarddeviatie van de drie groepen uitrekenen
print(np.sqrt(np.var(Age_business_class)))
print(np.sqrt(np.var(Age_eco_plus)))
print(np.sqrt(np.var(Age_eco)))
```

```python
#Dan gaan we aan de slag met de ANOVA
from scipy.stats import f_oneway
stats.f_oneway(random_subset['Age'][random_subset['Class'] == 1],
               random_subset['Age'][random_subset['Class'] == 2],
               random_subset['Age'][random_subset['Class'] == 3])
```

```python
#Dat is dus niet significant; saai, dus een andere ANOVA
stats.f_oneway(random_subset['Flight_Distance'][random_subset['Class'] == 1],
               random_subset['Flight_Distance'][random_subset['Class'] == 2],
               random_subset['Flight_Distance'][random_subset['Class'] == 3])
```

```python
#Eerst nog wat andere libraries om de Anova op een andere wijze en met andere output te doen
import statsmodels.api as sa
import statsmodels.formula.api as sfa
import scikit_posthocs as sp
```

```python
#Anova op een andere wijze:
lm = sfa.ols('Age~ C(Class)', data=random_subset).fit()
anova = sa.stats.anova_lm(lm)
print(anova)
```

```python
#Als er een significant verschil is tussen de drie groepen, weten we nog niet welke groepen verschillen
#t toetsen moeten dan duidelijkheid brengen, maar we hebben dan een correctie nodig
# We kiezen hier voor Tukey's correctie

import statsmodels.stats.multicomp as mc

comp = mc.MultiComparison(random_subset['Age'], random_subset['Class'])
post_hoc_res = comp.tukeyhsd()
post_hoc_res.summary()

```

```python
comp = mc.MultiComparison(random_subset['Flight_Distance'], random_subset['Class'])
post_hoc_res = comp.tukeyhsd()
post_hoc_res.summary()
```

```python
#Een andere veelgebruikte post hoc test is de t test met Bonferroni correctie
import statsmodels.stats.multicomp as mc

comp = mc.MultiComparison(random_subset['Flight_Distance'], random_subset['Class'])
tbl, a1, a2 = comp.allpairtest(stats.ttest_ind, method= "bonf")

tbl
```

```python
#Omdat er zoveel mogelijk is, nog een andere methode: Sidak's correctie
comp = mc.MultiComparison(random_subset['Flight_Distance'], random_subset['Class'])
tbl, a1, a2 = comp.allpairtest(stats.ttest_ind, method= "sidak")

tbl
```

```python
#Ook nu weer een alternatieve benadering
sp.posthoc_ttest(random_subset, val_col='Age', group_col='Class', p_adjust='bonferroni')
```

```python
#De verschillen kunnen ook geplot worden
post_hoc_res.plot_simultaneous(ylabel= "Class", xlabel= "Score Difference")
```

```python
#Eigenlijk moeten we eerst nog andere testen doen, voordat we een ANOVA mogen doen
#De eerste is de vraag in hoeverre de data normaal verdeeld is. Uitgangspunt is een 
#een test maar ook een visuele check. Bij grote steekproeven wordt immers elk klein
#verschil al significant. (De Catch22 is evenwel dat bij grote steekproeven de 
#centrale limietstelling geldt en derhalve de verdeling niet uitmaakt)
#Groot aantal testen; meest gebruikte  Shapiro-Wilk test, Kolmogorov-Smirnov (K-S) test
#We kiezen voor K-S test

import pandas as pd
import scipy.stats as stats
import statsmodels.formula.api as smf

model = smf.ols("Age ~ C(Class)", data= random_subset).fit()

stats.kstest(model.resid, 'norm')

```
