## Verover de data

Download `wineQualityN.csv` vanaf Brightspace, laad deze in een pandas DataFrame.

```python
df = pd.read_csv('wineQualityN.csv')
```

Bekijk de data.

```python
df
```

Observeer dat één kolom tekstlabels bevat. Vervang deze kolom door een kolom met dezelfde informatie gecodeerd in cijfers.

```python
# Retrieve all unique wine types and create an encode and decode dict.
wine_types = list(df['type'].unique())
wine_types.sort()
wine_types_encode = {t: i for i, t in enumerate(wine_types)}
wine_types_decode = {i: t for i, t in enumerate(wine_types)}
print(wine_types_encode)

# Encode all wine types in the type column
wine = df.copy()
wine['type'] = [wine_types_encode[t] for t in df['type']]
wine
```

Bepaal de correlaties tussen de kleur van de wijn en alle andere features.

```python
wine.corr()[['type']].sort_values(['type'], ascending=False)
```

## Visualiseer de data om inzicht te krijgen

Kies de vier features met de hoogste correlatie (hint: deze kan negatief zijn) en plot de correlaties tussen deze features (hint: een scatter matrix).

```python
# Create a correlation list only containing the correlations with type.
highest_correlation = wine.corr()[['type']]
# Make sure to also take in account negative correlations.
highest_correlation = highest_correlation.abs()
# Sort the list to have the highest correlation on top.
highest_correlation = highest_correlation.sort_values(['type'], ascending=False)
# Take the highest correlating features including type.
highest_correlation = highest_correlation.iloc[:5]
print(highest_correlation)

import seaborn as sns
%matplotlib inline

# Create a pair plot of the features with the highest correlation.
sns.set()
sns.pairplot(df[highest_correlation.index].dropna(), hue='type', height=4, palette={'white': 'gray', 'red': 'red'});
```

Kijk naar de histogrammen op de diagonaal. Wat kun je opmerken over de kleur?

- Red wine has less sulfur dioxide than white wine.
- White wine has less volatile acidity than red wine.
- White wine has less chlorides than red wine
- White wine has less fixed acidity than red wine's mean. Red wine is a bit more spread out on this dimension.

You can clearly see though that the individual features are insufficient for sepparating the red from the white wines.

Wat valt je op aan overige correlatieplots?

There are some clear clusters forming inbetween all the plots having the `total sulfur dioxide` feature.

## Prepareer de data voor Machine Learning algoritmes.

Zijn er rijen met ontbrekende data?

```python
# Count the missing fields per column.
wine.isna().sum()
```

Zorg dat er geen ontbrekende waarden meer zijn.

```python
# Drop the non existing data.
wine.dropna(inplace=True)
```

We verwijderen de kolom `quality`, deze is niet zinvol voor onze classificatie.

```python
wine = wine.drop(['quality'], axis = 1)
```

We willen nu een classifier maken die bepaalt welk kleur een wijn heeft op basis van de andere features.

Maak een feature matrix X en een target vector y. 

```python
X = wine[wine.columns[1:]]
```

```python
y = wine[wine.columns[0]]
```

Splits de data in een test en een training set, zet hierbij de `random_state` op 42

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Selecteer een model, stel het model in en train het model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_model = model.predict(X_test)
print("Accuracy", accuracy_score(y_test, y_model))
```

## Verbeter je model

Probeer twee andere classifiers op de data.

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
y_model = model.predict(X_test)
print("Accuracy", accuracy_score(y_test, y_model))
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_model = model.predict(X_test)
print("Accuracy", accuracy_score(y_test, y_model))
```

Classificeer de data nu met een neuraal netwerk.

```python
# Keras verwacht een matrix (array van array's) als output:
y_train = y_train.values.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
```

One Hot encode de labels

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)

print(y_train[:10])
```

Maak nu een netwerk met 4 lagen.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D

model = Sequential()
model.add(Dense(50, activation='tanh', input_shape=(11,)))
model.add(Dense(2, activation='softmax'))
```

Compileer het netwerk en genereer een samenvatting.

```python
from keras.optimizers import Adam

model.compile(Adam(lr=0.0003),'categorical_crossentropy',metrics=['accuracy'])

model.summary()
```

Train het model. Begin met 10 epochs.

```python
model.fit(X_train, y_train, epochs=10)
```

Evalueer het model.

```python
score = model.evaluate(X_test, y_test)

print('Score:', score[1])
```

Lets try to train another decision tree!

```python
# Reprepare train and test data.
X = wine[wine.columns[1:]]
y = wine[wine.columns[0]]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a list of features the model needs to train against.
features = [
    'total sulfur dioxide',
    'chlorides',
    'volatile acidity',
    'sulphates',
    'density',
    'residual sugar',
    'fixed acidity',
    'free sulfur dioxide'
]

print("Feature list size: ", len(features))

model = DecisionTreeClassifier()
model.fit(X_train[features], y_train)
y_model = model.predict(X_test[features])
print("Accuracy", accuracy_score(y_test, y_model))
```

This optimization results in a higher accuracy (previously: 0.9870). But do take in account that this model might cause fluctuating accuracies per training round.
