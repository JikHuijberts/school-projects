#### Removing/editing columns

There are multiple functions to alter data with pandas voor nu zijn deze functies het handigst en gebruikt
- ``drop()``
- ``fillna()``
- ``describe()``


```python
'''Columns verwijderen'''
data.drop(columns=["Name"])

'''NaN/na vullen. median() wordt gebruikt voor het midden te berekenen.
Dit kan ook mean() gebruiken voor het gemiddelde'''
data["Age"] = data["Age"].fillna(data["Age"].median())

'''describe() voor meer data of pre-gegenereerde data'''
data.describe()
```

#### Group By
``groupby([item])`` kan gebruikt worden voor het groeperen van data per geselecteerde object. Dit is handig om dingen te vergelijken per item die geselecteerd is.

```python
'''Voorbeeld van group by'''
data[["Fare", "Pclass"]].groupby(["Pclass"]).describe()
```
