```python
# imports
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy.stats as stats
```

Read the csv and save it as a variable

```python
data = pd.read_csv("Airlinesatisfaction.csv")
```

Show data

```python
data.head()
```

Prepare data by removing the non sensical values

```python
data = data.drop(columns = ["Unnamed: 0","id"])
data.head()
```

Make the data easly readable

```python
data['satisfaction'].replace({'neutral or dissatisfied': 0, 'satisfied': 1},inplace = True)
data['Gender'].replace({'Male': 0, 'Female': 1}, inplace = True)
data['Customer Type'].replace({'disloyal Customer': 0, 'Loyal Customer': 1}, inplace = True)
data['Type of Travel'].replace({'Personal Travel': 0, 'Business travel': 1}, inplace = True)
data['Class'].replace({'Eco': 1, 'Eco Plus': 2, 'Business':3}, inplace = True)
```

Create variables for the long texts

```python
fd ="Flight Distance"
ddim = "Departure Delay in Minutes"
adim = "Arrival Delay in Minutes"
```

#####  Test  normaalverdeling

```python
from scipy.stats import kstest, norm
def print_kstest(val):
    ks_stats, p_value = kstest(data[val], "norm")
    print("ks result: " + str(ks_stats) + " P value: " +  str(p_value))
```

Test age

```python
print_kstest("Age")
```

Test flight distance

```python
print_kstest(fd)
```

Test Departure Delay in Minutes

```python
print_kstest(ddim)
```

Test Arrival Delay in Minutes

```python
print_kstest(adim)
```

##### Get Z indexes

Scaler kan de data naar een z-index veranderen.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
z_data = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
```

##### Heatmap and scatter matrix

Met een heatmap kan je makkelijk de correlatie van bepaalde waarden aflezen.

```python
plt.figure(figsize = (20, 20))
sns.heatmap(data.corr(), annot = True, square = True)
```

Heatmap z-scores

```python
plt.figure(figsize = (20, 20))
sns.heatmap(z_data.corr(), annot = True, square = True)
```

#### Scatter matrix 

Een scatter matrix laat de verspreiding van de data goed zien. Dit kan gecombineerd worden met een andere column om te kijken of er een relatie tussen de data is.

```python
scatter_matrix(data[["Age",fd,adim,ddim]], figsize = (12, 8))
plt.show()
```

```python
scatter_matrix(z_data[["Age",fd,adim,ddim]], figsize = (12, 8))
plt.show()
```
