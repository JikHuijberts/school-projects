```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import tree
import matplotlib.pyplot as plt
```

## Dataset

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = pd.read_csv("zoo.data")
data.columns = ["animal_name","hair","feathers","eggs","milk","airborne","aquatic","predator","toothed","backbone",
                "breathes","venomous","fins","legs","tail","domestic","catsize","type"]
```

## Analyse Probleem
In deze dataset zijn er verschillende typen dieren gecategoriseerd in groepen. Hieruit probeer ik het type dier uit
kenmerken te vinden. De data had nog geen columns. Dit is in principe geen probleem, maar voor de overzicht is het wel
handig om dit erbij te hebben voor duidelijkheid.

### Dimensie dataset

```python
data.shape
```

```python
data.head(5)
```

Zoals hierboven te zien is, zijn er weinig rijen data. Daarentegen zijn er wel veel features. Dit moet vooral helpen om
de verschillende data te onderscheiden

### Soorten data/features

Bijna alle waarden zijn 0/1 dit komt omdat het uit kenmerken van een dier komt. Daarom zijn er niet veel stappen nodig
om de data in te zien.

```python
sns.factorplot('type', data=data,kind="count", aspect=2)
```

```python
# heatmap to show correlations
plt.subplots(figsize=(20,15))
ax = plt.axes()
ax.set_title("Correlation Heatmap")
corr = data.corr()
sns.heatmap(corr, annot=True,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

### Welke preprocessing stappen zijn nodig?
Er zijn verschillende pre-processing stappen nodig. De naam van het dier moet weg. De reden hiervoor is dat het model
niet moet trainen op de naam en de directe connectie daartussen, maar naar een connectie tussen de eigenschappen en
daardoor het type. (zoals zoogdier)

### Classifiers
Ik ga een decision tree gebruiken voor het duidelijk voorspellen van de data. <br/>
Daarnaast wordt er gekeken naar linear regression. Na encoding zijn de waarden 1 t/m 9 en kunnen in principe linear
gezien worden. Daarom wil ik kijken of hier ook accurate waarden uit komen.

Ik verwacht dat de Decision tree het heel goed gaat doen vanwege de vele features en de 1/0 antwoorden daarop
hieruit lijkt het vooral afstrepen.

```python
x_data = data.drop(columns={"type","animal_name"})
y_data = data["type"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data,y_data, train_size=0.8, random_state=3)
```

```python
dtc = tree.DecisionTreeRegressor()
dtc.fit(X_train,y_train)
print(dtc.score(X_test, y_test))

lrc = LinearRegression()
lrc.fit(X_train, y_train)
print(lrc.score(X_test,y_test))
```

## Resultaten
De resultaten zijn heel goed voor de Decision Tree en de Linear regression.
Hiervoor heb ik ook de DecisionTreeClassifier class gebruikt maar die gaf een resultaat van precies 0.9.
Voor de linear regression ziet het er ook heel goed uit. Sinds de dataset niet bepaald groot is ben ik verrast van het
resultaat.

#### Decision Tree Plot
Een van de voordelen van de decision tree is dat de waardes geplot kunnen worden en gevisualiseerd.

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtc)
```

## Evaluatie/conclusie
Doordat de data heel duidelijk is zijn er hoge resultaten. Het is lastig om data hiervoor te plotten en duidelijk
onderzoek over de zoo data te vinden. Hierdoor was het meer kijken wat er gebeurt en daarna op reageren. Dit keer kwam
het goed uit en is de Decision Tree erbovenuit gekomen.
